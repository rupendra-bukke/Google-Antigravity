## 1. DataFrame: The "Spreadsheet Creator" üìâ

A **DataFrame** is just a fancy word for a table. Rows and Columns. That's it.
We can build one from a Dictionary (remember those?).

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [24, 30, 22, 35],
    "City": ["New York", "Paris", "London", "Tokyo"],
    "Power_Level": [100, 200, 150, 9000]  # Over 9000?!
}

df = pd.DataFrame(data)

# Look at it! It's beautiful.
print(df)
```

## 2. Inspection: The "Magnifying Glass" üîç

When you have 1 million rows, you can't just `print(df)`. You need to peek at it.
- `.head()`: Shows the first 5 rows.
- `.info()`: Tells you about data types (int, float, object/string).
- `.describe()`: Quick stats (mean, max, min).

```python
# Just the tip (of the iceberg)
print("--- HEAD ---")
print(df.head(2))

# The Stats
print("\n--- SPECS ---")
print(df.describe())
```

## 4. Rows: The "Laser Pointer" üî¶

Selecting rows is tricky. We have two lasers:
- `.loc`: Select by **Label** (Index name).
- `.iloc`: Select by **Integer Position** (0, 1, 2...).

```python
# Get the row at index 1 (Bob)
bob_row = df.iloc[1]
print(bob_row)
```

### üèãÔ∏è‚Äç‚ôÄÔ∏è Practice Task: The Sniper

## 5. Filtering: The "VIP List" üéüÔ∏è

Just like NumPy, we can filter rows based on conditions.
"Show me everyone older than 25!"

```python
# The Condition
mask = df["Age"] > 25

# applying the mask
older_folks = df[mask]

print(older_folks)
```

### üèãÔ∏è‚Äç‚ôÄÔ∏è Practice Task: The Top Tier
1. Filter `shop_df` to find items where "Price" is **greater than 100**.
2. Print them.

```python
# Your code here:
expensive_items = shop_df[shop_df["Price"] > 100]
print(expensive_items)
```

## 6. Aggregation: The "Scoreboard" üèÜ

Pandas is great at math. You can sum, average (mean), count, min, max whole columns instantly.

```python
total_power = df["Power_Level"].sum()
average_age = df["Age"].mean()

print(f"Total Power: {total_power}")
print(f"Average Age: {average_age}")
```

### üèãÔ∏è‚Äç‚ôÄÔ∏è Practice Task: The Stat Master
1. Calculate the **mean** (average) "Price" of items in `shop_df`.
2. Calculate the **sum** of "Stock".
3. Print both.

```python
# Your code here:
avg_price = shop_df["Price"].mean()
total_stock = shop_df["Stock"].sum()

print(f"Average Price: ${avg_price}")
print(f"Total Stock: {total_stock}")
```
