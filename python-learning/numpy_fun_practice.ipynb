## 1. Arrays: The "Lego Baseplate" üß±

A Python list is a bag of random stuff. A NumPy array is a **Lego Baseplate**. Everything must be the *same type* (usually numbers), and that makes it super fast to build on.

```python
# Creating a 1D Array (A single line of legos)
arr_1d = np.array([1, 2, 3, 4, 5])
print(f"1D Array: {arr_1d}")

# Creating a 2D Array (A grid of legos)
arr_2d = np.array([
    [1, 2, 3], 
    [4, 5, 6]
])
print(f"\n2D Array values:\n{arr_2d}")
print(f"Shape (Rows, Cols): {arr_2d.shape}")
```

## 2. Operations: The "Magic Wand" ‚ú®

In Python, if you want to multiply every item in a list by 2, you need a loop (boring!).
In NumPy, you just wave your wand (multiply the array), and it happens to **everyone at once**.
This is called **Broadcasting**.

```python
prices = np.array([10, 20, 30])

# THE MAGIC: Discount everything by 50% instantly
discount_prices = prices * 0.5

print(f"Original: {prices}")
print(f"Discounted: {discount_prices}")
```

## 3. Slicing: The "Cookie Cutter" üç™

Just like Python strings, you can slice arrays. But with 2D arrays (tables), you can slice rows and columns independently.
It's like using a cookie cutter to grab just the chocolate chips.

```python
grid = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Grab the single element at Row 0, Column 1 (Remember 0 is the start)
print(f"Element at (0,1): {grid[0, 1]}")

# Grab the FIRST ROW only
print(f"Row 0: {grid[0, :]}")

# Grab the LAST COLUMN only (using : to say 'all rows')
print(f"Last Column: {grid[:, 2]}")
```

## 4. Filtering: The "Bouncer v2.0" üïµÔ∏è

We can ask questions to the array, like "Who here is greater than 50?"
The array returns a mask of Trues and Falses, which we can use to filter.

```python
ages = np.array([12, 18, 20, 15, 60])

# Ask the question
is_adult = ages >= 18
print(f"The Truth: {is_adult}")

# Filter the crowd
adults_only = ages[is_adult]
print(f"Adults Only: {adults_only}")
```

## 5. Reshaping: The "Transformer" ü§ñ

Sometimes you have a long line of numbers, but you need a square. NumPy lets you reshape data without changing the values.
It's like taking a deck of cards spread out and stacking them up.

```python
# A long line of 8 numbers
line = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# Transform into 2 Rows and 4 Columns
matrix = line.reshape(2, 4)
print(f"Reshaped:\n{matrix}")

# TRANSFORM AGAIN! 4 Rows, 2 Columns
tall_matrix = line.reshape(4, 2)
print(f"\nReshaped Again:\n{tall_matrix}")
```
