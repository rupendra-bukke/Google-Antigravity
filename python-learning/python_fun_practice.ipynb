## 2. Strings: The "Word Salad" ü•ó

Strings are just text. But python lets us chop, slice, and dice them like a chef with a grudge.

```python
my_text = "Tim is cool"

# TRICK: The "Moonwalk" (Reverse a string)
# The [::-1] slice means "start at the end, go to the beginning, step back by 1"
reversed_text = my_text[::-1]

print(f"Normal: {my_text}")
print(f"Backwards: {reversed_text}")

# Shouting contest
print(my_text.upper() + "!!!")
```

## 4. Loops: "Groundhog Day" üîÑ

Loops let you do the same thing over and over without going crazy writing the code 100 times.

```python
# The "Chore Wheel" loop
chores = ["Dishes", "Vacuum", "Feed the Dragon", "Laundry"]

print("--- Honey, I'm doing chores! ---")
for task in chores:
    print(f"Task: {task}... DONE! ‚úÖ")
    if task == "Feed the Dragon":
        print("   (Phew, didn't get eaten!) üî•")

print("--- All done! ---")
```

## 5. Functions: The "Spell Book" üßô‚Äç‚ôÇÔ∏è

Instead of writing the same code (incantation) everywhere, we write it once in our Spell Book (define a function) and just shout the name when we need it.

```python
def summon_coffee(cups, sugar=True):
    """Summons coffee from the void."""
    order = f"‚òï {cups} cups of dark matter java"
    if sugar:
        order += " with sweet crystals"
    else:
        order += " (bitter like my soul)"
    return order

# Casting the spell
print(summon_coffee(1, sugar=False))
print(summon_coffee(3, sugar=True))
```

## 8. Dictionaries: The "Spy's Dossier" üìÇ

Lists are great, but sometimes you need to know *what* something is, not just where it is in line.
A **Dictionary** (dict) is like a Spy's Dossier. You look up a codename (Key) to find the secret info (Value).

```python
# Creating a dossier
spy_dossier = {
    "Codename": "007",
    "Name": "James Bond",
    "Weapon": "Walther PPK",
    "Martinis_Consumed": 42
}

# Accessing info by Key (Not by number!)
print(f"Agent Name: {spy_dossier['Name']}")

# Adding new intel
spy_dossier["Status"] = "Active"

# Updating intel
spy_dossier["Martinis_Consumed"] = 43

print(spy_dossier)
```

## 10. Sets: The "VIP Club" üö´

A **Set** is a collection where *duplicates are banned*. If you try to add the same thing twice, it just ignores you.
Also, sets have no order. It's a chaotic mosh pit of unique items.

```python
# Creating a set with curly braces {}
vip_list = {"Alice", "Bob", "Charlie", "Alice", "Alice"}

# Notice "Alice" only appears once!
print(f"VIP List: {vip_list}")

# Checking if someone is in the club
print(f"Is Bob invited? {'Bob' in vip_list}")
```

## 11. Error Handling: The "Safety Net" üï∏Ô∏è

Sometimes code crashes. It happens. Instead of letting the whole program explode, we use a **Try/Except** block to catch the error safely.

```python
try:
    # Asking for trouble
    result = 10 / 0
except ZeroDivisionError:
    print("üö® WHOA! You can't divide by zero! The universe will implode. üö®")
    result = 0

print(f"Safe Result: {result}")
```
